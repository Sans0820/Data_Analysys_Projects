```python
import pandas as pd

data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Amit", "Sadhya", "Diksha", "Neha", "Pratik"],
    "Department": ["HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 60000, 55000, 70000, 65000],
    "Experience": [3, 4, 2, 5, 4]
}

df = pd.DataFrame(data)

while True:
    print("\n1. Highest Paid")
    print("2. Average Salary by Department")
    print("3. Search Employee")
    print("4. Add Employee")
    print("5. Delete Employee")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    match choice:

        case 1:
            highest = df[df["Salary"] == df["Salary"].max()]
            print("\nHighest Paid Employee")
            print(highest)

        case 2:
            print("\nAverage Salary by Department")
            print(df.groupby("Department")["Salary"].mean())

        case 3:
            name = input("Enter employee name: ")
            result = df[df["Name"].str.lower() == name.lower()]

            if result.empty:
                print("Employee not found.")
            else:
                print(result)

        case 4:
            empid = int(input("Enter Employee ID: "))
            name = input("Enter Name: ")
            dept = input("Enter Department: ")
            salary = int(input("Enter Salary: "))
            exp = int(input("Enter Experience: "))

            new_row = {
                "EmployeeID": empid,
                "Name": name,
                "Department": dept,
                "Salary": salary,
                "Experience": exp
            }

            df.loc[len(df)] = new_row
            print("Employee Added Successfully!")
            print(df)

        case 5:
            empid = int(input("Enter Employee ID to delete: "))

            if empid in df["EmployeeID"].values:
                df = df[df["EmployeeID"] != empid]
                print("Employee Deleted Successfully!")
                print(df)
            else:
                print("Employee ID not found.")

        case 6:
            print("Exiting...")
            break

        case _:
            print("Invalid Choice")
```

```python
#Highest Paid Employee
highest = df[df["Salary"] == df["Salary"].max()]
print(highest)
```

```python
#Lowest Paid
lowest = df[df["Salary"] == df["Salary"].min()]
print(lowest)
```

```python
#Employee Count in Each Department
print(df["Department"].value_counts())
```

```python
#Most Experienced Employee
print(df[df["Experience"] == df["Experience"].max()])
```

```python
#Least Experienced Employee
print(df[df["Experience"] == df["Experience"].min()])
```

```python
#Employees with Salary Greater Than a Given Amount
salary = int(input("Enter salary: "))
print(df[df["Salary"] > salary])
```

```python
#Export Data
df.to_csv("employees.csv", index=False)
```
