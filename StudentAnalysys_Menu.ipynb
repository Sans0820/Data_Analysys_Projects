```python
import pandas as pd

data = {
    "RollNo":[1,2,3,4,5],
    "Name":["Asha","Ravi","Karan","Neha","Priya"],
    "Marks":[80,65,90,75,88],
    "City":["Mumbai","Pune","Delhi","Mumbai","Pune"]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
topper=df[df["Marks"]==df["Marks"].max()]
print(topper)
```

```python
print(df["Marks"].mean())
```

```python
print(df[df["Marks"]>80])
```

```python
print(df[df["City"]=="Mumbai"])
```

```python
def grade(marks):
  if marks>=90:
    return "A+"
  elif marks>=80:
    return "A"
  elif marks>=70:
    return "B"
  else:
    return "C"
df["Grade"]=df["Marks"].apply(grade)
print(df)
```

```python
print(df["City"].value_counts())
```

```python
print(df.groupby("City")["Marks"].mean())
```

```python
df["FinalMarks"] = df["Marks"] + 5
print(df)
```

```python
df.to_csv("student_result.csv", index=False)
```

**Menu Driven Program**

```python
new_df = pd.read_csv("student_result.csv")
```

```python
print(new_df)
```

```python
def grade(marks):
  if marks>=90:
    return "A+"
  elif marks>=80:
    return "A"
  elif marks>=70:
    return "B"
  else:
    return "C"
df["Grade"]=df["Marks"].apply(grade)
while True:
  choice = int(input("Enter choice: "))
  match choice:
    case 1:
      print(new_df)
    case 2:
      print("Toper =",df[df["Marks"]==df["Marks"].max()])
    case 3:
      print("Average marks=",df['Marks'].mean())
    case 4:
      stname=input("Enter student name")
      print(df[df["Name"]==stname])
    case 5:
      roll = int(input("Enter Roll No: "))
      name = input("Enter Name: ")
      marks = int(input("Enter Marks: "))
      city = input("Enter City: ")

      # Calculate grade and final marks for the new student
      new_grade = grade(marks)
      new_final_marks = marks + 5

      df.loc[len(df)] = [roll, name, marks, city, new_grade, new_final_marks]

      print("Student added successfully!")
      print(df)
    case 6:
      df.to_csv("students.csv", index=False)
      print("Data saved successfully")
    case 7:
      break




```
